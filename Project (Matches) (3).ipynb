```python
cat_cols=matches.select_dtypes(include=['object']).columns
num_cols=matches.select_dtypes(include=np.number).columns.tolist()
print("Categorial Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
for col in num_cols:
    print(col)
    print('Skew:',round(matches[col].skew(),2))
    plt.figure(figsize=(15,4))
    plt.subplot(1,2,1)
    matches[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1,2,2)
    sns.boxplot(x=matches[col])
    plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')

# Plot each categorical variable
sns.countplot(ax=axes[0, 0], x='city', data=matches, color='blue', 
              order=matches['city'].value_counts().index);
sns.countplot(ax=axes[0, 1], x='player_of_match', data=matches, color='blue', 
              order=matches['player_of_match'].value_counts().index);
sns.countplot(ax=axes[1, 0], x='team1', data=matches, color='blue', 
              order=matches['team1'].value_counts().index);
sns.countplot(ax=axes[1, 1], x='team2', data=matches, color='blue', 
              order=matches['team2'].value_counts().index);
sns.countplot(ax=axes[2, 0], x='toss_winner', data=matches, color='blue', 
              order=matches['toss_winner'].head(20).value_counts().index);
sns.countplot(ax=axes[2, 1], x='toss_decision', data=matches, color='blue', 
              order=matches['toss_decision'].head(20).value_counts().index);
# Rotate and adjust font size of labels for better readability
for ax in axes.flatten():
    ax.tick_params(axis='x', labelrotation=90, labelsize=10)

# Adjust layout to maximize visibility
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust bottom margin to make room for title
plt.show()
```
