```python
cat_cols=delivery.select_dtypes(include=['object']).columns
num_cols = delivery.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
for col in num_cols:
    print(col)
    print('Skew :', round(delivery[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    delivery[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=delivery[col])
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')

# Plot each categorical variable
sns.countplot(ax=axes[0, 0], x='fielder', data=delivery, color='blue', 
              order=delivery['fielder'].value_counts().index);
sns.countplot(ax=axes[0, 1], x='bowling_team', data=delivery, color='blue', 
              order=delivery['bowling_team'].value_counts().index);
sns.countplot(ax=axes[1, 0], x='batsman', data=delivery, color='blue', 
              order=delivery['batsman'].value_counts().index);
sns.countplot(ax=axes[1, 1], x='non_striker', data=delivery, color='blue', 
              order=delivery['non_striker'].value_counts().index);
sns.countplot(ax=axes[2, 0], x='bowler', data=delivery, color='blue', 
              order=delivery['bowler'].head(20).value_counts().index);
sns.countplot(ax=axes[2, 1], x='player_dismissed', data=delivery, color='blue', 
              order=delivery['player_dismissed'].head(20).value_counts().index);

# Rotate labels for better readability
axes[1, 1].tick_params(labelrotation=45)
axes[2, 0].tick_params(labelrotation=90)
axes[2, 1].tick_params(labelrotation=90)

# Adjust layout to maximize visibility
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust bottom margin to make room for title
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')

# Plot each categorical variable
sns.countplot(ax=axes[0, 0], x='fielder', data=delivery, color='blue', 
              order=delivery['fielder'].value_counts().index);
sns.countplot(ax=axes[0, 1], x='bowling_team', data=delivery, color='blue', 
              order=delivery['bowling_team'].value_counts().index);
sns.countplot(ax=axes[1, 0], x='batsman', data=delivery, color='blue', 
              order=delivery['batsman'].value_counts().index);
sns.countplot(ax=axes[1, 1], x='non_striker', data=delivery, color='blue', 
              order=delivery['non_striker'].value_counts().index);
sns.countplot(ax=axes[2, 0], x='bowler', data=delivery, color='blue', 
              order=delivery['bowler'].head(20).value_counts().index);
sns.countplot(ax=axes[2, 1], x='player_dismissed', data=delivery, color='blue', 
              order=delivery['player_dismissed'].head(20).value_counts().index);

# Rotate and adjust font size of labels for better readability
for ax in axes.flatten():
    ax.tick_params(axis='x', labelrotation=90, labelsize=10)

# Adjust layout to maximize visibility
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust bottom margin to make room for title
plt.show()
```
